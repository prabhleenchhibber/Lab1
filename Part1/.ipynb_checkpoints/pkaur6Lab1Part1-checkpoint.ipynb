#PROBLEM 2
Bar plot with lines indicating values for sales1 and sales2.
Sales1 is simply using c() to create a path witht he values we explicitly mention.
Sales2 however uses rpois() function which takes an upper limit and a lower limit and then plots with random values.
Upperlimit-45 , lower limit-12 in our case.

```r
sales1<-c(12,14,16,29,30,31,35,37,45)
sales2<-rpois(12,45)
par(bg="white")
plot(sales1, col="blue",type="o", ylim=c(0,100), xlab="Month", ylab="Sales")
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny =NULL)
legend("topright",inset=.05,c("sales1","sales2"), fill=c("blue","red"),horiz=TRUE)
```

#PROBLEM 3
read.table("txtfilename") is used to read data present in a text file .
We are plotting box on the basis of values present with us and calculating median and displaying it.

```r
sales<- read.table("salesdata.txt",header=T)
#barplot(as.matrix(sales),main="Sales Data" , ylab="Total",beside=T,col=rainbow(7))
fn<-boxplot(sales,col=c("orange","green"))$stats
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

#PROBLEM 4
I am taking data from 2 csv files in next example and the using them to plot a histogram and a plot with lines indicating values.
Plot with lines : Blue line represents Facebook and Red one represents Apple. We can see the path and the trend followed by FB and Apple for days wrt frequency.
Histogram: It is just depicting the data from apple file and we have specifically told the compiler to use rainbow colors rather than default. we have Adj Close column on x axis and Frequency on y axis.

```r
fb1<-read.csv("FB.csv")
aapl1<-read.csv("AAPL.csv")
par(bg="white")
plot(aapl1$Adj.Close,col="blue",pch=22,type="o",xlab="Days",ylab="Price")
     #,ylim=c(0,100),xlab="Days",ylab="Price")
lines(fb1$Adj.Close,type="o",pch=22,lty=2,col="red")
legend("topright",inset=.05,c("Facebook","Apple"),fill=c("blue","red"),horiz=TRUE)
hist(aap1$Adj.Close, col=rainbow(8))
```

#PROBLEM 6
In next example we plotting US map using google APIs and then plotting points with respect to places mentioned in visited variable.
geocode() gives  the lattitude and longitude required to plot the points in map.

```r
library(ggmap)
library(maptools)
library(maps)
register_google(key = Sys.getenv("GoogleAKey"))
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima Peru", "Johannesbury, SA")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)
```

```r
library("ggmap")

library("maptools")

library(maps)
register_google(key = Sys.getenv("GoogleAKey"))
visited <- c("Miami,FL","SFO", "New York", "Buffalo", "Dallas, TX")

ll.visited <- geocode(visited)

visit.x <- ll.visited$lon

visit.y <- ll.visited$lat

map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0))

points(visit.x,visit.y, col="yellow", pch=36)

```
